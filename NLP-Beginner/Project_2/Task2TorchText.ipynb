```python
import torch
import pandas as pd
import numpy as np
import torchtext
from torchtext import data,datasets
from torchtext.vocab import GloVe
from torchtext import data
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import random
import spacy
import gensim.downloader as api
from gensim.models import TfidfModel
from gensim.corpora import Dictionary

spacy_en = spacy.load('en_core_web_sm')

def tokenizer(text): # create a tokenizer function
    return [tok.text for tok in spacy_en.tokenizer(text)]

TEXT = data.Field(sequential=True, tokenize=tokenizer, lower=True, fix_length=100)
LABEL = data.Field(sequential=False, use_vocab=False)
```

```python
train_data = data.TabularDataset(
        path='./train.tsv',format='tsv',
        skip_header=True,
        fields = [('PhraseId', None),('SentenceId', None),('Phrase', TEXT),('Sentiment',LABEL)])
test_data = data.TabularDataset(
        path='./test.tsv',format='tsv',
        skip_header=True,
        fields = [('PhraseId', None),('SentenceId', None),('Phrase', TEXT)])
```

```python
#vector.unk_init = init.xavier_uniform
#构建字典和batch
TEXT.build_vocab(train_data, vectors=GloVe(name='6B', dim=300))
LABEL.build_vocab()
#划分数据集
train_data, valid_data = train_data.split(random_state = random.seed(10))



#print(train_data[0].Phrase)
BATCH_SIZE = 16
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
train_iterator, valid_iterator = data.BucketIterator.splits(
    (train_data, valid_data), 
    sort_key=lambda x: len(x.Phrase), 
    batch_size = BATCH_SIZE,
    device = device)



train_on_gpu=torch.cuda.is_available()
if(train_on_gpu):
    print('Training on GPU.')
else:
    print('No GPU available, training on CPU.')


#train_iterator.__dict__.keys()
```

```python
for idx, batch in enumerate(train_iterator):
    print(batch)
    text, label = batch.Phrase, batch.Sentiment
    print(text.shape, label.shape)
```

```python
class SentimentRNN(nn.Module):

    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_size,n_layers, drop_prob=0.2):
        super(SentimentRNN, self).__init__()
        self.vocab_size=vocab_size
        #self.input_size=input_size
        self.hidden_dim=hidden_dim
        self.output_size=output_size
        self.embedding_dim = embedding_dim
        self.n_layers=n_layers
        #随机生成Embedding
        #self.word_embeddings = nn.Embedding(vocab_size, embedding_dim) 
        # 若使用预训练的词向量，需在此处指定预训练的权重
        #使用预训练的文件Embedding
        self.word_embeddings = nn.Embedding.from_pretrained(TEXT.vocab.vectors)
        #self.word_embeddings.weight.data.copy_(TEXT.vocab.vectors)
        self.word_embeddings.weight.data.requires_grad=False
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, n_layers,dropout=drop_prob)  
        self.dropout = nn.Dropout(drop_prob)
        self.decoder = nn.Linear(hidden_dim, output_size)

    def forward(self, sentence):
        embeds = self.word_embeddings(sentence)
        lstm_out = self.lstm(embeds)[0]  
        # 取最后一个时间步
        #print(lstm_out.size())
        mean_token_output = torch.mean(lstm_out, dim=0)  # 针对第1维进行mean 的平均操作
        #print(mean_token_output.size())
        #final = lstm_out[-1]  
        y = self.decoder(mean_token_output)   
        return y
```

```python
vocab_size = int(TEXT.vocab.vectors.shape[0]) # +1 for the 0 padding
output_size = 5
embedding_dim = 300
hidden_dim = 512
n_layers = 2

net = SentimentRNN(vocab_size,embedding_dim, hidden_dim,output_size, n_layers)

print(net)
# loss and optimization functions
optimizer = optim.Adam(filter(lambda p: p.requires_grad, net.parameters()), lr=0.001)
loss_funtion = F.cross_entropy
```

```python
epochs = 3 # 3-4 is approx where I noticed the validation loss stop decreasing

counter = 0
print_every = 100
clip=5 # gradient clipping

# move model to GPU, if available
if(train_on_gpu):
    net.cuda()

net.train()
losses=[]
# train for some number of epochs
for epoch, batch in enumerate(train_iterator):
    optimizer.zero_grad()
    #start = time.time()
    predicted = net(batch.Phrase)
    #print(predicted.size())
    loss = loss_funtion(predicted, batch.Sentiment)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
    
    if epoch % print_every == 0:
        print("Epoch: {}/{}...".format(epoch+1,len(train_iterator)),
              "Loss: {:.6f}...".format(loss.item()),
              "Val_loss: {:.6f}...".format(np.mean(losses)))
```

```python
net.train()
valid_losses=[]
num_correct = 0
# train for some number of epochs
for epoch, batch in enumerate(valid_iterator):
    optimizer.zero_grad()
    #start = time.time()
    predicted = net(batch.Phrase)

    valid_loss = loss_funtion(predicted, batch.Sentiment)
    valid_losses.append(valid_loss.item())
    #loss.backward()
    #optimizer.step()
    _, pred = torch.max(predicted,1)
    # Get test data loss and accuracy
    correct_tensor = pred.eq(batch.Sentiment.view_as(pred))
    correct = np.squeeze(correct_tensor.numpy()) if not train_on_gpu else np.squeeze(correct_tensor.cpu().numpy())
    num_correct += np.sum(correct)

# -- stats! -- ##
# avg test loss
print("Test loss: {:.3f}".format(np.mean(valid_losses)))

# accuracy over all test data
test_acc = num_correct/len(valid_iterator.dataset)
print("Test accuracy: {:.3f}".format(test_acc))
```
